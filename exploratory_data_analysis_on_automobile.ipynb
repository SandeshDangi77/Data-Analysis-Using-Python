```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

**Data Loading**

```python
df_automobile = pd.read_csv('/content/drive/MyDrive/23CB045/Automobile_data.csv')
```

**Data Cleaning**
*  Data contains "?" replace it with NAN


```python
df_data = df_automobile.replace('?',np.nan)
df_data.isnull().sum()
```

**Missing Data**
*  Fill missing data of  *normalised-losses, price, horsepower, peak-rpm, bore, stroke* with the respective column mean
*  Fill missing data category Number of doors with the mode of the column i.e. Four

```python
df_temp = df_automobile[df_automobile['normalized-losses']!='?']
normalised_mean = df_temp['normalized-losses'].astype(int).mean()
df_automobile['normalized-losses'] = df_automobile['normalized-losses'].replace('?',normalised_mean).astype(int)

df_temp = df_automobile[df_automobile['price']!='?']
normalised_mean = df_temp['price'].astype(int).mean()
df_automobile['price'] = df_automobile['price'].replace('?',normalised_mean).astype(int)

df_temp = df_automobile[df_automobile['horsepower']!='?']
normalised_mean = df_temp['horsepower'].astype(int).mean()
df_automobile['horsepower'] = df_automobile['horsepower'].replace('?',normalised_mean).astype(int)
```

```python
df_temp = df_automobile[df_automobile['peak-rpm']!='?']
normalised_mean = df_temp['peak-rpm'].astype(int).mean()
df_automobile['peak-rpm'] = df_automobile['peak-rpm'].replace('?',normalised_mean).astype(int)

df_temp = df_automobile[df_automobile['bore']!='?']
normalised_mean = df_temp['bore'].astype(float).mean()
df_automobile['bore'] = df_automobile['bore'].replace('?',normalised_mean).astype(float)

df_temp = df_automobile[df_automobile['stroke']!='?']
normalised_mean = df_temp['stroke'].astype(float).mean()
df_automobile['stroke'] = df_automobile['stroke'].replace('?',normalised_mean).astype(float)
```

```python

df_automobile['num-of-doors'] = df_automobile['num-of-doors'].replace('?','four')
df_automobile.head()
```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
# Define the date range
start_date = datetime(2022, 1, 1)
end_date = datetime(2024, 12, 31)
date_range = (end_date - start_date).days
df_automobile['Date'] = [start_date + timedelta(days=np.random.randint(0, date_range)) for _ in range(len(df_automobile))]
print(df_automobile.head())
```

**HERE I DONT HAVE A TIME DATA FOR TIME SERIES SO I M GOING TO ADD SOME ATTRIBUTES TO MY DATASET**


```python
import numpy as np

# Convert price to numeric, handling missing values if present
df_automobile['price'] = pd.to_numeric(df_automobile['price'], errors='coerce')

# Generate random quantities between 1 and 10
df_automobile['Quantity'] = np.random.randint(1, 11, size=len(df_automobile))

# Calculate Total Sales
df_automobile['Total_Sales'] = df_automobile['Quantity'] * df_automobile['price']

# Drop rows where Total_Sales or Date values are NaN due to missing prices
df_automobile.dropna(subset=['Total_Sales'], inplace=True)

# Display the first few rows to verify the additions
print(df_automobile[['price', 'Quantity', 'Total_Sales']].head())
```

**TIME SERIES ANALYSIS**:

```python
import matplotlib.pyplot as plt

df_automobile['Date'] = pd.to_datetime(df_automobile['Date'])
df_automobile.set_index('Date', inplace=True)

# Resample the data to get monthly total sales
monthly_sales = df_automobile['Total_Sales'].resample('M').sum()

# Plot the monthly total sales
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales, marker='o', color='b')
plt.title("Monthly Total Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```

This plot will show you the total sales trend over each month. Look for patterns like seasonality (repeating trends over certain months), increases, or decreases in sales over time.

```python
# Calculate a 3-month moving average for smoother trends
monthly_sales_ma = monthly_sales.rolling(window=3).mean()

# Plot the original monthly sales and the 3-month moving average
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales, marker='o', color='b', label='Monthly Sales')
plt.plot(monthly_sales_ma, marker='o', color='red', label='3-Month Moving Average')
plt.title("Monthly Total Sales with 3-Month Moving Average")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.legend()
plt.grid(True)
plt.show()
```

**Findings**
Monthly Sales Volatility: The blue line shows high
variability month-to-month, with some very sharp peaks and drops. This could indicate some seasonality or external factors influencing sales spikes.

.Moving Average Smoothing: The 3-month moving average (red line) smooths out some of this volatility, allowing us to see a more general trend over time. The smoothed line helps identify whether the overall trend is increasing, decreasing, or stable.
