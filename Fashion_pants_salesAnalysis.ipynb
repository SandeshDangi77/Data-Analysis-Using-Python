```python
df.drop_duplicates(keep='first',inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

```python

# Visualization 1: Histogram of Sales Quantity
plt.figure(figsize=(10, 5))
sns.histplot(df['Sales_Qty'], bins=20, kde=True, color='skyblue')
plt.title("Distribution of Sales Quantity")
plt.xlabel("Sales Quantity")
plt.ylabel("Frequency")
plt.show()

# Visualization 2: Box Plot of Total Sales
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Total_Sales', color='lightgreen')
plt.title("Box Plot of Total Sales")
plt.xlabel("Total Sales")
plt.show()
```

```python
# Set up the figure size for all plots
plt.figure(figsize=(10, 6))

# 1. Histogram for Sales Quantity (Distribution)
plt.subplot(2, 2, 1)
sns.histplot(df['Sales_Qty'], kde=True, color='skyblue')
plt.title("Distribution of Sales Quantity")
plt.xlabel("Sales Quantity")
plt.ylabel("Frequency")

# 2. Linear Plot for Total Sales over time (Assuming Date column exists)
plt.subplot(2, 2, 2)
df['Date'] = pd.to_datetime(df['Date'])  # Ensure 'ODate' is in datetime format
sns.lineplot(x='Date', y='Total_Sales', data=df, color='green')
plt.title("Total Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales")

# 3. Bar Plot for Average Sales Quantity by Product
plt.subplot(2, 2, 3)
sns.barplot(x='Product_Name', y='Sales_Qty', data=df, palette='Blues_d')
plt.title("Average Sales Quantity by Product")
plt.xlabel("Product Name")
plt.ylabel("Average Sales Quantity")

# 4. Box Plot for Unit Price (to identify the spread and any outliers)
plt.subplot(2, 2, 4)
sns.boxplot(data=df, x='Unit_Price', color='lightcoral')
plt.title("Box Plot for Unit Price")
plt.xlabel("Unit Price")

# Show all plots
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Unit_Price', y='Sales_Qty', color='green')
plt.title("Scatter Plot: Sales Quantity vs. Unit Price")
plt.xlabel("Unit Price")
plt.ylabel("Sales Quantity")
plt.show()
```

```python
# Compute the correlation matrix
correlation_matrix = df[['Sales_Qty', 'Unit_Price', 'Total_Sales']].corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix: Sales Data")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Convert 'Order_Date' to datetime format if it's not already
df['Date'] = pd.to_datetime(df['Date'])

# Aggregate total sales by month
monthly_sales = df.resample('M', on='Date')['Total_Sales'].sum()

# Plotting the time series
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linestyle='-', color='b')
plt.title("Total Sales Over Time (Monthly)")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```
